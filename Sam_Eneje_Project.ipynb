# BLOG

```
%%html
<img src='/nbextensions/Adjacency Matrix- Blog.png'/ alt="Image" height="500" width="500">
```

```
# Reading file for analysis - BLOG
G = nx.read_graphml('/content/Adjacency Matrix- Blog.graphml', node_type=str, edge_key_type=int)

# This code is just for checking with xlsx file of Adjacency Matrix
print("Node Degree")
for v in G:
    print('%s %s' % (v, G.degree(v)))
```

```
# How many nodes we have at BLOG enviroment?
G.number_of_nodes()
```

```
# How many edges we have at BLOG enviroment?
G.number_of_edges()
```

```
# Efficiency Network Metric in % value at BLOG enviroment
Eff = (G.number_of_edges()/(G.number_of_nodes()*G.number_of_nodes()))*100
round(Eff, 3)
```

```
print("Closeness Centrality")
c = nx.closeness_centrality(G)
for v in G.nodes():
    print((v, round(c[v],4)))
```

```
print("Betweenness")
b = nx.betweenness_centrality(G)
for v in G.nodes():
    print((v, round(b[v],4)))
```

```
print("Degree Centrality")
d = nx.degree_centrality(G)
for v in G.nodes():
    print((v, round(d[v],4)))
```
